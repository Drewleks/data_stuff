```
# Создадим новую колонку с датами в правильном типе

data['date'] = pd.to_datetime(data['Местное время в Москве (ВДНХ)'], dayfirst=True)
```

```
data
```

```
# Косяк №3: данные даны задом наперед, метеослужба вначале держит новые записи, а потом старые
# Из-за этого когда мы завтра будем строить тренд, он нарисуется "в прошлое", а в будущее
# Отсортируем данные в правильно порядке
data.sort_values('date')  # Отсортировать данные по колонке date ПО ВОЗРАСТАНИЮ (это новый объект)
data = data.sort_values('date')  # Отсортировать данные по колонке date ПО ВОЗРАСТАНИЮ (обновили перемнную data)
```

```
data
```

```
# Косяк номер 5: при сортировке развернулся индекс - сбросим его
# data.reset_index()  # новый объект
data = data.reset_index() # индекс идет правильно от 0 до 1
```

```
# Фильтрация данных
start_date = pd.Timestamp(year=2019, day=1, month=1)
condition = data['date'] > start_date
data[condition]
```

```
data_short = data[condition]

x = data_short['date']
y = data_short['T']
plt.plot(x, y)
```
